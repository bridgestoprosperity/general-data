# Cleaning path data
one of the major cleaning operations is going through all the paths that go from an h3 index to the same destination. I'll be keeping the longest path

```python
def keep_longest_path(df):
    df = df.copy()
    print("Original df shape:", df.shape)
    
    # Drop rows with None in destination_coords
    df = df.dropna(subset=['destination_coords'])
    print("After dropping None coords:", df.shape)
    
    df['coords_str'] = df['destination_coords'].apply(lambda x: str(x.tolist()) if hasattr(x, 'tolist') else str(x))
    df['path_length'] = df['exit_point_path'].str.len()
    
    # Drop rows where exit_point_path is None/NaN (which would make path_length NaN)
    df = df.dropna(subset=['path_length'])
    print("After dropping NaN path lengths:", df.shape)
    
    idx = df.groupby(['h3_index', 'coords_str'])['path_length'].idxmax()
    # Drop any NaN indices
    idx = idx.dropna()
    
    result = df.loc[idx].drop(['path_length', 'coords_str'], axis=1)
    
    print("Result df shape:", result.shape)
    print("Rows removed:", df.shape[0] - result.shape[0])
    
    return result
```

```python
market_data_clean = keep_longest_path(market_data)
```

# Transforming Paths to Geospatial Data

```python
health_paths = pd.read_parquet(os.path.join(output_folder_path, "all_health_paths.parquet"))
```

```python
# Convert the exit_point_path column to LineString geometries
health_paths['geometry'] = health_paths['exit_point_path'].apply( lambda path: LineString(np.array([np.array(point) for point in path])) )
health_paths
```

```python
def paths_to_geodataframe(df):
   def safe_linestring(path):
       if len(path) < 2:  # Drop if 0 or 1 points
           return None
       else:
           return LineString(np.array([np.array(point) for point in path]))

   # Convert both columns, keeping only valid LineStrings
   df['exit_geometry'] = df['exit_point_path'].apply(safe_linestring)
   df['path_geometry'] = df['path_to_exit_point'].apply(safe_linestring)

   # Only keep rows where at least one geometry is valid
   mask = (df['exit_geometry'].notna()) | (df['path_geometry'].notna())
   df_filtered = df[mask].copy()

   # Join geometries (only LineStrings now)
   def safe_merge(row):
       geoms = [g for g in [row['path_geometry'], row['exit_geometry']] if g is not None]
       if len(geoms) == 1:
           return geoms[0]
       else:
           return linemerge(geoms)

   df_filtered['geometry'] = df_filtered.apply(safe_merge, axis=1)
   
   # Convert to GeoDataFrame
   return gpd.GeoDataFrame(df_filtered, geometry='geometry')
```

```python
gdf = paths_to_geodataframe(health_paths)
gdf = gdf.set_crs(epsg=4326, allow_override=True)
gdf = gdf.drop(columns=['exit_point_path', 'path_to_exit_point', 'exit_geometry', 'path_geometry'])
gdf.to_parquet(os.path.join(output_folder_path, "all_health_paths_geodataframe.parquet"), index=False)
```

```python
edu_paths = pd.read_parquet(os.path.join(output_folder_path, "all_education_paths.parquet"))
print("edu paths read")
gdf = paths_to_geodataframe(edu_paths)
print("edu paths converted to geodataframe")
gdf = gdf.set_crs(epsg=4326, allow_override=True)
gdf = gdf.drop(columns=['exit_point_path', 'path_to_exit_point', 'exit_geometry', 'path_geometry'])
print("writing edu paths to parquet")
gdf.to_parquet(os.path.join(output_folder_path, "all_education_paths_geodataframe.parquet"), index=False)
```
