# Estimating travel time saved
Answering the question how much travel time we predict is saved 

```python
import pandas as pd
import geopandas as gpd
import os
```

```python
rwanda = gpd.read_file("https://fikamap-web-app.s3.us-west-1.amazonaws.com/data/rwa_travel_time_hex-8-mini.geojson")
rwanda
```

```python
rwanda.columns
```

```python
# drop all columns that don't end in potential impact
rwanda = rwanda.drop(columns=[col for col in rwanda.columns if not col.startswith("time_delta")])
rwanda
```

```python
# if all columns are 0, drop the row
rwanda = rwanda.loc[~(rwanda.iloc[:, 1:] == 0).all(axis=1)]
rwanda
```

```python
rwanda.describe()
```

```python
def get_average_non_zero(column):
    """
    Get the average of a column, ignoring 0 values.
    """
    return rwanda.loc[rwanda[column] != 0, column].mean()
```

```python
average_list = []
for col in rwanda.columns[1:]:
    average_list.append(get_average_non_zero(col))
    print(f"{col}: {get_average_non_zero(col)}")
```

```python
# average of average list
average = sum(average_list) / len(average_list)
print(f"Average: {average}")
```
